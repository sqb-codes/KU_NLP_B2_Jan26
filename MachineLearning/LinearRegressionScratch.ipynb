```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("headbrain.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
X = df['Head Size(cm^3)'].values
y = df['Brain Weight(grams)'].values
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y)
plt.show()
```

```python
mean_x = np.mean(X)
mean_y = np.mean(y)
```

```python
mean_x, mean_y
```

```python
numer = 0
denom = 0
n = len(X)

for i in range(n):
    numer += (y[i] - mean_y) * (X[i] - mean_x)
    denom += (X[i] - mean_x) ** 2

slope = numer / denom
intercept = mean_y - slope * mean_x
print("Slope :",slope)
print("Intercept :",intercept)
```

```python
predictions = []
for i in range(n):
    y_pred = slope * X[i] + intercept
    predictions.append(int(y_pred))
```

```python
print("Prediction:",predictions[:5])
print("Actual:",y[:5])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y)
plt.plot(X, predictions, 'r')
plt.show()
```

```python
error = 0
for i in range(n):
    error += (predictions[i] - y[i]) ** 2

mse = error / n
print("MSE :",mse)
print("RMSE :",np.sqrt(mse))
```

# Gradient Descent

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X = np.reshape(X, (-1,1))
y = np.reshape(y, (-1,1))

minmax = MinMaxScaler()
scaled_x = minmax.fit_transform(X)
scaled_y = minmax.fit_transform(y)
```

```python
scaled_x[:5]
```

```python
def gradient_descent(epochs, alpha):
    slope, inter = 0,0
    for i in range(epochs):
        y_pred = scaled_x * slope + inter
        loss = y_pred - scaled_y
        gradSlope = (2/n) * np.dot(loss.T, scaled_x)
        gradInter = (2/n) * np.sum(loss)
        slope = slope - gradSlope * alpha
        inter = inter - gradInter * alpha
    return slope, inter
```

```python
epochs = 8000
alpha = 0.01
slope, inter = gradient_descent(epochs, alpha)
print("Slope :",slope)
print("Inter :", inter)
```

```python
predictions = []
for i in range(n):
    y_pred = slope[0][0] * scaled_x[i] + inter
    predictions.append(y_pred)

plt.figure(figsize=(10,6))
plt.scatter(scaled_x, scaled_y)
plt.plot(scaled_x, predictions, 'r')
plt.show()
```
