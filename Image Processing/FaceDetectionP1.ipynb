```python
# Face detection
import cv2
import numpy as np
```

```python
# Load Haar cascade for face detection
haar_features = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
# img = cv2.imread('msd.jpeg')
img = cv2.imread('team.jpg')
```

```python
# img
# scaleFactor - how much image size is reduced at each step
haar_features.detectMultiScale(img, 1.3)
```

```python
faces = haar_features.detectMultiScale(img, 1.3)
for (x,y,w,h) in faces:
    # img - original image on which we want to draw rectangle
    # (x,y) - (top-left corner) co-ordinate where face it detected
    # (x+w, y+h) - (bottom-right corner)
    # (0,255,0) - green color
    # 2 - thickness of border
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
```

```python
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
